```python
from math import log
import operator
```

```python
def shannonent(dataset):
    hangshu = len(dataset)  #记录数据行数，样本个数
    labelcount = {}    #装label个数
    for  featVec in dataset:
        currentlabel = featVec[-1] #选取label:Y值  这里的数据加载为一个矩阵，若是个frame好像不能这样子
        if currentlabel not in labelcount:  #若是还未构建，需要增加一个
            labelcount[currentlabel] += 1
        labelcount[currentlabel] += 1  #这些操作把label中Y类型储存在字典中
    shannonent = 0.0
    for key in labelcount:
        prob = float(labelcount[key])/hangshu  #计算不同类别的信息熵中的概率
        shannonent -= prob*log(prob,2)  #计算信息熵
    return shannonent
```

划分数据集

```python
def splitdataset(dataset,axis,value):
    retdataset = []
    for featVec in dataset:
        if featVec[axis] == value: #featvec为数据集的每一行,取每一行的第axis列，看是否等于value，axis和value都为在最后由遍历获得
            reducedfeatVec = featVec[:axis]
            reducedfeatVec.extend(featVec[axis+1:])#这两部把axis这列不要了
            retdataset.append(reducedfeatVec)
    return retdataset  #返回的是没有了axis这列的数据集  另外会将数据分割成n份（有多少个value就有多少份
```

```python
def choosebestfeature(dataset):
    n_feature = len(dataset[0]) - 1#样本数
    baseEntropy = shannonent(dataset)  #计算信息熵
    bestinfogain = 0.0 #信息增益
    bestfeature = -1 #最优特征的索引
    for i in range(n_feature):  #准备遍历所有特征
        #获取第i个特征
        featlist = [example[i] for example in dataset]  #取数据集中第i个特征的值
        uniquevals = set(featlist)  #去重，包含了i特征所有值的唯一值
        newentropy = 0.0
        for value in uniquevals:
            subdataset = splitdataset(dataset,i,value) #value为第i个特征中的取值
            prob = len(subdataset)/float(len(dataset))   #会的到每个value在样本中出现的概率
            newentropy += prob*shannonent(subdataset) #计算条件经验熵
        infogain = baseEntropy - newentropy  #计算信息增益
        print('第%d个特征的信息增益为: %.3f'%(i,infogain))
        if (infogain > bestinfogain):
            bestinfogain = nfogain
            bestfeature = i
    return bestfeature
```

```python
def majoritycnt(classlist):
    classcount = {}
    for vote in classlist:
        if vote not in classcount.keys():classcount[vote] = 0
        classcount[vote] += 1
    sortedclasscount = sorted(classcount.items(),key=operator.itemgetter(1),reverse = 0)
    return sortedclasscount[0][0]  #投票选出最终的分类
```

```python
def createtree(dataset,labels,featlabels):
    classlist = [example[-1] for example in dataset]
    if classlist.count(classlist[0]) == len(classlist):
        return classlist[0]  #如果标签都一样就不用划分了
    if len(dataset[0]) == 1 or len(labels) == 0: #所有特征都弄完了
        return majoritycnt(classlist)   #得到最终分类
    bestfeat = choosebestfeature(dataset)  #选出最优特征
    bestfeatlabel = labels[bestfeat]  #得到最优特征的名字
    featlabels.append(bestfealabel)
    mytree = {bestfeatlabel:{}}
    del(labels[bestfeat])  #删除已经使用过的标签
    featvalues = [example[bestfeat] for example in dataset]  #选取最佳特征中的值
    uniquevals = set(featvalues)
    for value in uniquevals:
        mytree[bestfeatlabel][value] = createtree(splitdataset(dataset,bestfeat,value),labels,featlabels)
    return mytree
    
```

```python
fr = open('lenses.txt')
dataset = [inst.strip().split('\t') for inst in fr.readlines()]
lensesLabels = ['age', 'prescript', 'astigmatic', 'tearRate']
featLabels = []
```

```python
lensesTree = createtree(dataset,lensesLabels,featLabels)
print(lensesTree)
```
